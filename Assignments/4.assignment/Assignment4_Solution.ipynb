### **1. Count the frequency of each vowel in a string**

**Question:**
Write a function that takes a string and returns a dictionary with the frequency of each vowel.

**Example:**
Input: `"Programming in Python"`
Output: `{'o': 2, 'a': 1, 'i': 2}`

```python
def vowel_frequency(s):
    vowels = "aeiou"
    freq = {}
    for char in s.lower():
        if char in vowels:
            freq[char] = freq.get(char, 0) + 1
    return freq

# Example
print(vowel_frequency("Programming in Python"))
```

### **2. Check if a string is a palindrome (ignoring case and spaces)**

**Example:**
Input: `"A man a plan a canal Panama"`
Output: `True`

```python
def is_palindrome(s):
    cleaned = s.replace(" ", "").lower()
    return cleaned == cleaned[::-1]

# Example
print(is_palindrome("A man a plan a canal Panama"))
```

### **3. Sum with next element in list**

**Question:**
Given a list of integers, return a new list where each element is the sum of itself and its next element. Last element remains unchanged.

**Example:**
Input: `[1, 2, 3, 4]`
Output: `[3, 5, 7, 4]`

```python
def sum_with_next(lst):
    result = []
    for i in range(len(lst)):
        if i < len(lst) - 1:
            result.append(lst[i] + lst[i + 1])
        else:
            result.append(lst[i])
    return result

# Example
print(sum_with_next([1, 2, 3, 4]))
```

### **4. Remove all occurrences of an element from a list**

**Example:**
Input: `([1, 2, 3, 2, 4], 2)`
Output: `[1, 3, 4]`

```python
def remove_element(lst, value):
    return [x for x in lst if x != value]

# Example
print(remove_element([1, 2, 3, 2, 4], 2))
```

### **5. Count unique elements using sets**

**Example:**
Input: `[1, 2, 2, 3, 4, 4, 4]`
Output: `4`

```python
def count_unique(lst):
    return len(set(lst))

# Example
print(count_unique([1, 2, 2, 3, 4, 4, 4]))
```

### **6. Concatenate two tuples and remove duplicates**

**Example:**
Input: `(1, 2, 3), (2, 3, 4)`
Output: `(1, 2, 3, 4)` (order may vary)

```python
def concat_unique(t1, t2):
    return tuple(set(t1 + t2))

# Example
print(concat_unique((1, 2, 3), (2, 3, 4)))
```

### **7. Most frequent character (excluding spaces)**

**Example:**
Input: `"hello world"`
Output: `'l'`

```python
def most_frequent_char(s):
    s = s.replace(" ", "")
    freq = {}
    for char in s:
        freq[char] = freq.get(char, 0) + 1
    return max(freq, key=freq.get)

# Example
print(most_frequent_char("hello world"))
```

### **8. Common elements using set (no loops)**

**Example:**
Input: `[1, 2, 3], [2, 3, 4]`
Output: `[2, 3]`

```python
def common_elements(lst1, lst2):
    return list(set(lst1) & set(lst2))

# Example
print(common_elements([1, 2, 3], [2, 3, 4]))
```

### **9. Check if all characters in a string are unique**

**Example:**
Input: `"python"`
Output: `True`

```python
def all_unique(s):
    return len(set(s)) == len(s)

# Example
print(all_unique("python"))
```

### **10. Maximum product of any two distinct numbers in a tuple**

**Example:**
Input: `(1, 3, -4, -6, 7)`
Output: `24`

```python
import itertools

def max_product(tup):
    max_prod = float('-inf')
    for a, b in itertools.combinations(tup, 2):
        prod = a * b
        if prod > max_prod:
            max_prod = prod
    return max_prod

# Example
print(max_product((1, 3, -4, -6, 7)))
```
