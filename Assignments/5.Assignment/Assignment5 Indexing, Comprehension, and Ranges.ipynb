# Assignment 5: Indexing, Comprehension, and Ranges

### 1. Rotate a list to the right by k positions using slicing  
* Input: [1, 2, 3, 4, 5], k = 2  
* Output: [4, 5, 1, 2, 3]

### 2. From a list of names, return a list of names that start and end with a vowel using list comprehension  
* Input: ['Alice', 'Eve', 'Oscar', 'Uma', 'Ian', 'Bob']  
* Output: ['Alice', 'Eve', 'Oscar', 'Uma', 'Ian']


### 3. Generate all numbers from 1 to 100 divisible by both 3 and 7 using a single line of code  
* Output: [21, 42, 63, 84]

### 4. Write a function that returns the squares of all even numbers in a list, but only if the square is less than 100  
* Input: [1, 2, 5, 8, 10, 12]  
* Output: [4, 64]

### 5. Create a list of all two-digit numbers that are palindromes using list comprehension  
* Output: [11, 22, 33, 44, 55, 66, 77, 88, 99]

### 6. Given a sentence, build a dictionary using dictionary comprehension where the keys are words and the values are their lengths  
* Input: "Build a dictionary of word lengths"  
* Output: {'Build': 5, 'a': 1, 'dictionary': 10, 'of': 2, 'word': 4, 'lengths': 7}

### 7. Generate a dictionary of ASCII characters from 'a' to 'z' using range() and dictionary comprehension  
* Output (example): {'a': 97, 'b': 98, ..., 'z': 122}

### 8. Given a list of tuples (name, score), create a dictionary from it using dictionary comprehension and exclude names with scores below 60  
* Input: [("Alice", 85), ("Bob", 55), ("Carol", 75)]  
* Output: {'Alice': 85, 'Carol': 75}

### 9. Extract the middle 3 elements from a list of at least 7 elements using slicing  
* Input: [10, 20, 30, 40, 50, 60, 70]  
* Output: [30, 40, 50]

### 10. Write a program that creates a 2D list (matrix) of size 3x3 with values filled using range()  
* Output:  
  * [[1, 2, 3],  
     [4, 5, 6],  
     [7, 8, 9]]
