### **1. Rotate a list to the right by `k` positions using slicing**

**Input:** `[1, 2, 3, 4, 5]`, `k = 2`
**Output:** `[4, 5, 1, 2, 3]`

```python
def rotate_right(lst, k):
    k = k % len(lst)  # handle cases where k > length
    return lst[-k:] + lst[:-k]

# Example
print(rotate_right([1, 2, 3, 4, 5], 2))
```

### **2. Filter names that start and end with a vowel using list comprehension**

**Input:** `['Alice', 'Eve', 'Oscar', 'Uma', 'Ian', 'Bob']`
**Output:** `['Alice', 'Eve', 'Oscar', 'Uma', 'Ian']`

```python
def names_start_end_vowel(names):
    vowels = 'aeiouAEIOU'
    return [name for name in names if name[0] in vowels and name[-1] in vowels]

# Example
print(names_start_end_vowel(['Alice', 'Eve', 'Oscar', 'Uma', 'Ian', 'Bob']))
```

### **3. Generate numbers from 1 to 100 divisible by both 3 and 7 (one-liner)**

**Output:** `[21, 42, 63, 84]`

```python
divisible_by_3_and_7 = [x for x in range(1, 101) if x % 3 == 0 and x % 7 == 0]
print(divisible_by_3_and_7)
```

### **4. Squares of even numbers, only if square < 100**

**Input:** `[1, 2, 5, 8, 10, 12]`
**Output:** `[4, 64]`

```python
def squares_under_100(lst):
    return [x**2 for x in lst if x % 2 == 0 and x**2 < 100]

# Example
print(squares_under_100([1, 2, 5, 8, 10, 12]))
```

### **5. All two-digit palindromes using list comprehension**

**Output:** `[11, 22, 33, 44, 55, 66, 77, 88, 99]`

```python
palindromes = [x for x in range(10, 100) if str(x) == str(x)[::-1]]
print(palindromes)
```

### **6. Word lengths in a sentence using dictionary comprehension**

**Input:** `"Build a dictionary of word lengths"`
**Output:** `{'Build': 5, 'a': 1, 'dictionary': 10, 'of': 2, 'word': 4, 'lengths': 7}`

```python
def word_lengths(sentence):
    return {word: len(word) for word in sentence.split()}

# Example
print(word_lengths("Build a dictionary of word lengths"))
```

### **7. Dictionary of ASCII characters from 'a' to 'z'**

**Output (partial):** `{'a': 97, 'b': 98, ..., 'z': 122}`

```python
ascii_dict = {chr(i): i for i in range(ord('a'), ord('z') + 1)}
print(ascii_dict)
```

### **8. Dictionary from (name, score) tuples excluding score < 60**

**Input:** `[("Alice", 85), ("Bob", 55), ("Carol", 75)]`
**Output:** `{'Alice': 85, 'Carol': 75}`

```python
def filter_scores(pairs):
    return {name: score for name, score in pairs if score >= 60}

# Example
print(filter_scores([("Alice", 85), ("Bob", 55), ("Carol", 75)]))
```

### **9. Extract the middle 3 elements from a list**

**Input:** `[10, 20, 30, 40, 50, 60, 70]`
**Output:** `[30, 40, 50]`

```python
def middle_three(lst):
    mid = len(lst) // 2
    return lst[mid - 1:mid + 2]

# Example
print(middle_three([10, 20, 30, 40, 50, 60, 70]))
```

### **10. Create a 3x3 matrix filled with values using `range()`**

**Output:**

```
[[1, 2, 3],
 [4, 5, 6],
 [7, 8, 9]]
```

```python
matrix = [[i for i in range(row, row + 3)] for row in range(1, 10, 3)]
print(matrix)
```
